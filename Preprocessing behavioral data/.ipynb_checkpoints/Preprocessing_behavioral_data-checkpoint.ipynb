## Experiment representation in data files:

txt files contain animal (x,y) trajectories; orientation; stimulus dot position and size; episode code.

One **file** contains data from one experiment defined as one video recording.

One **experiment** consists of many repeated trials.

One **trial** consists of 4 episodes: 

**Episode** code:

* 0: 10 seconds before dot presentation - no stimulus, white background.
* 1: 0.5 seconds stimulus dot(s) presentation.
* 2: 59.5 seconds after dot presentation - no stimulus, white background.
* 3: 20 seconds radial sine grating.

**Stimuli** are defined as a set of many different stimuli and are presented in randomly shuffled order

Stimuli are one or two dots covering all possible combinations of these features:

+ position: 90 degrees from center of mass
+ size: big
+ side*: left/ right

*both dots are always shown on the same side for now

### Structure of the txt file:

rows represent frames

column labels:

X an0, Y an0, ori an0, X dot1, Y dot1, X dot2, Y dot2  --- these 7 columns repeat for each animal

+4 columns at the end - definition of stimulus size, side and episode:
(all animals see the same stimulus pattern

size dot1, size dot2, side code, episode code

** size indicates dot diameter in each frame ((need to check what the units mean))**

below, using a hard-coded frame (3 frames after stimulus start) to read of small vs large dot as size value 1 vs 2

** side code**
* left=1
* right=0

** total number of columns: (n_animals x 7) + 4 **

# Read raw data from txt files

+ populate list with matries containing raw trajectories
+ read txt file row by row. 
+ Check if row is shorter than 6 characters: exception!
+ Unhandled tracking errors result in a line 'none' without position for any animal.
+ replace such rows with adequate number of NaN

** truncate data after the last complete trial: **
experiments may have been aborted mid trial.

determine trial length (in frames) and use integer multiple number of frames.

### Sanity check that all trials are the same length

```python
def readTxtsToMat(fileList):
    matAll=[] # list for Matrices containing raw trajectory data of one experiment
    goodExperiment=[]

    for currTxt in fileList: # loop over txt files

        with open(currTxt) as f:
            mat=pd.read_csv(currTxt,delim_whitespace=True,header=None).values
        # determine trial length by checking where episode code changes to 0 (marks the beginning of a trial).
        TrialLen=int(np.median(np.diff(np.where((mat[:-1,-3]!=mat[1:,-3]) * ((mat[:-1,-3])==0)))))
        completeTrials=int(np.floor(mat.shape[0]/float(TrialLen)))
        tmp = np.diff(np.where((mat[:-1,-3]!=mat[1:,-3]) * ((mat[:-1,-3])==0)))
        goodExperiment.append(np.unique(tmp).shape[0]==1) # only have one unique trial length

        print ('experiment:',currTxt,'median trial length: ',TrialLen,' number of complete trials: ', completeTrials)
        mat=mat[:completeTrials*TrialLen,:]
        matAll.append(mat)

    #Sanity check that all trials are the same length
    if sum(goodExperiment)==len(goodExperiment): #all experiments are good (True)
        TrialLen=np.unique(tmp)[0]
        print ('OK. all experiments have the same trial length')
    else:
        TrialLen=0
        print ('problem, trials are not the same length, consider fixing!!')    

    return matAll,TrialLen
```

# Consolidate data into one DataFrame

final df is essentially a stacked version of raw data matrix. Each df row represents only one animal at one frame.

**data matrix format: repeating pattern of 7 columns**

animalX, animalY, orientation, point1X, point1Y, point2X, point12Y

**Additionally, the last four columns are shared for all animals:**

point1Size, point2Size, leftSide?, episode(see above for episode definition)

pointSize is the size actually displayed for each frame.

from it, calculate 's1' and 's2', representing a code for the entire trial how big the dot was. This is read from the second frame after dot presentation begins. At that frame, small dots have size 1 and large dots have size 2.

from size, compute a 'condition' code that labels the 8 possible size combinations 1-8

add 'trial' column to indicate trial number

```python
def matToDf(matAll,TrialLen,trajectoryLen=25,startOffset=300):

    print ('assembling dataFrame')
    numAnimals=9 # per experiment
    size_code_frame=2 #fist frame where c1/c2 reflect size/contrast code - hard code for now


    dfRaw=pd.DataFrame(columns=['x','y','o','xp1','yp1','xp2','yp2','s1b','s2b','l','e','c1','c2'])

    experiment=0
    for mat in matAll:
        

        TrialBegin=mat[:,-3]==99
        TrialBegin=TrialBegin[:-1*TrialLen]
        TrialBegin_f=np.where(TrialBegin)[0]+startOffset
        completeTrials=np.sum(TrialBegin)
        tmp = np.diff(np.where(TrialBegin))
        
        lastTrialBegin=np.where(TrialBegin)[-1][-1]
        
        pullFrames=np.repeat(TrialBegin_f,trajectoryLen)+np.tile(np.arange(trajectoryLen),completeTrials)

        #list of frames where dot size (small/large) can be read conveniently from raw data as (1/2)
        size_code_index=np.arange(size_code_frame,pullFrames.shape[0],trajectoryLen)        

        for an in range(numAnimals):



            animalColumns=range(an*7,7*(an+1))
            animalColumns=list(animalColumns)
            animalColumns.extend(range(mat.shape[1]-6,mat.shape[1]))
            dft=pd.DataFrame(mat[pullFrames,:][:,animalColumns],columns=['x','y','o','xp1','yp1','xp2','yp2','s1b','s2b','l','e','c1','c2'])
            
            dft['s1']=np.repeat(dft.iloc[size_code_index].s1b.values,trajectoryLen)
            dft['s2']=np.repeat(dft.iloc[size_code_index].s2b.values,trajectoryLen)
            dft['c1']=np.repeat(dft.iloc[size_code_index].c1.values,trajectoryLen)
            dft['c2']=np.repeat(dft.iloc[size_code_index].c2.values,trajectoryLen)
            dft['condition']=3*dft.s1.values+dft.s2.values#+(dft.s1.values+dft['c1'])+(dft.s2.values*dft['c2']) # generate a unique stimulus code defined by the combination of sizes
            dft['expAnimal']=an # animal identifyer looping within one experiment.
            dft['animalID']=an+(experiment*numAnimals) # animal identifyer counting over all experiments.
            dft['trial']=np.repeat(np.arange(completeTrials),trajectoryLen) # trial within one experiment
            dft['experiment']=experiment
            dft['frame']=np.tile(np.arange(trajectoryLen),completeTrials)
            dft['frameCont']=pullFrames
            
            
            dfRaw=pd.concat([dfRaw,dft])
        experiment+=1


    
    print ('length of all frames after truncation to', trajectoryLen, ' frames per trial:',dfRaw.shape[0])
    # populate treatment column using animalID and treatment code provided above.
    dfRaw['treatment']=treatment[dfRaw.animalID.values.astype('int')]
    stimCode=np.unique(dfRaw.condition,return_inverse=True)[1]

    return dfRaw
```

```python
def getData(fileList,trajectoryLen=25,startOffset=300):
    print ('reading txt files')
    matAll,TrialLen=readTxtsToMat(fileList)
    print ('-done-')
    df=matToDf(matAll,TrialLen,trajectoryLen=trajectoryLen,startOffset=startOffset)
    return df
```

## Align trajectories
### Move and rotate trajectories for same orientation and position when dot first appears

assembling a nested list:

traAll: list of all trials of one animal with same stimulus
traAllAll[i]: above for all stimuli
traAllAllAll[an]: above for all animals

```python
#helper functions to convert cartesian coordinates to polar and back. Used to rotate trajectories around origin
def cart2pol(x, y):
    theta = np.arctan2(y, x)
    rho = np.hypot(x, y)
    return theta, rho
def pol2cart(theta, rho):
    x = rho * np.cos(theta)
    y = rho * np.sin(theta)
    return x, y
```

```python
def rotateTrajectories(df, trajectoryLen=25):
    
    dotStart=0
    totalTrials=df.shape[0]/trajectoryLen

    startIndex=df.frame==dotStart
    startX=np.repeat(df.x[startIndex].values,trajectoryLen)
    startY=np.repeat(df.y[startIndex].values,trajectoryLen)
    startO=np.repeat(df.o[startIndex].values,trajectoryLen)
    df['xOriginal']=df.x
    df['yOriginal']=df.y

    x=df[['x','xp1','xp2']]-np.tile(startX,(3,1)).T
    y=df[['y','yp1','yp2']]-np.tile(startY,(3,1)).T
    traT,traR=cart2pol(x.values,y.values)
    traTn=(traT-np.tile(startO,(3,1)).T)+np.tile((np.pi/2.),(3,1)).T
    x,y=pol2cart(traTn,traR)
    
    t1=(-2*df.l+1).astype(int)
    t1=np.tile(t1,(3,1)).T
    x=t1*x
    df[['x','xp1','xp2']]=x
    df[['y','yp1','yp2']]=y
    
    df['front']=1*(df.y>0) # animal position forwards relative to dot start.
    df['right']=1*(df.x>0) # animal position right relative to dot start.
    df['left']=1*(df.x<0) # animal position left relative to dot start.
    r,t=cart2pol(df.x.values,df.y.values)
    df['r']=r #r indicates rho - angle from origin to current position.
    df['centerDist']=t #distance from origin.
    return df
```

# Fix rare tracking errors

rarely, the tracker jumps from animal to some piece of dirt.

Identify such frames by finding position jumps greater than threshold.

** remove entire trial when error occured **

```python
def fixTrackingErrors(df,trajectoryLen=25,jump_tresh=50):
       

    #compute steps as difference in distance from center between consecutive frames
    df['lastStep']=np.hstack([0,np.diff(df.centerDist.values)])

    #remove artifactual steps between consecutive trials
    #remove nan (presumably due to missing data)
    a=np.hstack([0,df.frame.values[1:]<df.frame.values[:-1]]).astype('bool')
    b=np.isnan(df.lastStep.values)
    ind= np.logical_or(a,b)
    df.loc[ind,'lastStep']=0

    #remove trajectory data for entire trial when large jump was detected
    ind=np.abs(df.lastStep.values)>jump_tresh
    
    #unique animal/trial combinations where jump occured

    anTrial=df[['animalID','trial']][ind].drop_duplicates() #tuple
    for an,tr in anTrial.values:
        #an=df.iloc[rows,:].animalID
        #tr=df.iloc[rows,:].trial
        id2=(df.animalID==an)&(df.trial==tr)
        #df.loc[id2,['x','y']=np.nan
        df.loc[id2,['x','y','centerDist','r','front','right','left']]=np.nan
        #print 'removed animal',an,'trial',tr

    n_trials_total=ind.shape[0]/float(trajectoryLen+1)
    removed=anTrial.shape[0]
    print ('removed a total of',removed , 'out of',n_trials_total,'animal trials (',100.*removed/n_trials_total,'%)')
    
    return df
```

```python
# plot escape endpoints as polar histogram.
def polarHist(x, y, **kwargs):
    ax = plt.gca()
    data = kwargs.pop("data") # get the current plot data
    if data.shape[0]>0: # attempt plot only if there is any data
    
        N = 20 #number of bins
        bins = np.linspace(0.0, 2*np.pi, N+1, endpoint=True) # histogram bin edges spanning full circle 0 - 2x pi
        arr=data.r.values #pull data
        arr=arr[~np.isnan(arr)] #exclude nan
        arr=arr+np.pi #add pi to transform range from (-pi,pi) to (0,2xpi) for transformation to (0-360) degrees.
        
        radii, tick = np.histogram(arr, bins = bins,density =True,normed=True)
        
        width = (2*np.pi) / N
        x_plot=bins[:-1]
        col=['k','m'][data.treatment.values[0]==1] #color based on treatment. This works when data is split using hue
        plt.bar(x_plot, radii, width=width, color=col,alpha=0.6)
```

```python
def meanAngle_d(x):
    x=np.array(x)
    theta=np.deg2rad(x)
    rho=np.ones((theta.shape[0],1))
    x,y=pol2cart(theta,rho)
    return np.degrees(cart2pol(np.nanmean(x),np.nanmean(y))[0])

def medianAngle_d(x):
    x=np.array(x)
    theta=np.deg2rad(x)
    rho=np.ones((theta.shape[0],1))
    x,y=pol2cart(theta,rho)
    return np.degrees(cart2pol(np.median(x),np.median(y))[0])
```

```python
#for each animal, determine the average escape endpoint.
meanEscape_animal=d.groupby(['condition','treatment','animalID']).mean()[['x','y']]
meanEscape_animal=meanEscape_animal.reset_index()

#calculate mean escape angle for each group
meanEscape_animal['escapeAngle']=cart2pol(meanEscape_animal.x,meanEscape_animal.y)[0]
meanEscape_animal['escapeAngle_deg']=np.mod(np.degrees(meanEscape_animal['escapeAngle'])+90,360)-180
meanEscape_animal['escapeAngle_deg_noRot']=np.degrees(meanEscape_animal.escapeAngle)

#Next, average over animals
meanEscape_condition=meanEscape_animal.groupby(['condition','treatment']).mean()[['x','y']]
meanEscape_condition=meanEscape_condition.reset_index()


meanEscape_conditionUnitLength=meanEscape_animal.groupby(['condition','treatment']).escapeAngle_deg_noRot.apply(meanAngle_d)
meanEscape_conditionUnitLength=meanEscape_conditionUnitLength.reset_index()
meanEscape_conditionUnitLength['escapeAngle']=np.radians(meanEscape_conditionUnitLength.escapeAngle_deg_noRot)

#calculate mean escape angle for each group
meanEscape_condition['escapeAngle']=cart2pol(meanEscape_condition.x,meanEscape_condition.y)[0]
meanEscape_condition['escapeAngle_deg']=np.degrees(meanEscape_condition['escapeAngle'])
#meanEscape_condition
meanEscape_conditionUnitLength
```

```python
def plotStimulus(x, y, **kwargs):
    scode=[0,1.26,1.86]
    #ax = plt.gca()
    data = kwargs.pop("data")
    if data.shape[0]>0:
        trialList=data.trial.unique()
        for t in trialList:
            dc=data[data.trial==t]
            if dc.trial.values[0]<17:
                x=dc.xp1.values[0]
                y=dc.yp1.values[0]
                ss=dc.s1.values[0]
                s=int(1.6*15*scode[int(ss)]/2.)
                currCol=(1-dc.c1.values[0])*np.array([1,1,1])
                circle1 = plt.Circle((x, y), s, color=currCol)
                ax.add_artist(circle1)
                x=dc.xp2.values[0]
                y=dc.yp2.values[0]
                ss=dc.s2.values[0]
                s=int(1.6*15*scode[int(ss)]/2.)
                currCol=(1-dc.c2.values[0])*np.array([1,1,1])
                circle1 = plt.Circle((x, y), s, color=currCol)
                ax.add_artist(circle1)

            x=dc.x.values
            y=dc.y.values
            plt.plot(x[-1], y[-1],  'ok',ms=10, alpha=1)
```

```python
def plotTra(x, y, **kwargs):
    ax = plt.gca()
    data = kwargs.pop("data")
    trialList=data.trial.unique()
    for t in trialList:
        #print t
        data[data.trial==t].plot(x='x',y='y',ax=ax,alpha=1,ls='--', lw=3)
    
```

```python
animalID_to_plot=0
g = sns.FacetGrid(dfEarly[(dfEarly.condition==7)&(dfEarly.animalID==animalID_to_plot)],hue='trial',
                  height=10,
                  sharex=True, sharey=True, despine=False)


sns.set_style('white')
g.map(plt.axvline, x=0, ls=":", c=".5")
g.map(plt.axhline, y=0, ls=":", c=".5")
print ('plotting')
#g.map(plt.plot,'x', 'y')
g.map_dataframe(plotTra, 'x', 'y');
g.map_dataframe(plotStimulus, 'x', 'y');
plt.subplots_adjust(top=0.95)
g.fig.suptitle('escape trajectories for one fish',fontsize=16)
for ax in g.axes.flat:
    ax.set_xlim([-325,325])
    ax.set_ylim([-325,325])
```
